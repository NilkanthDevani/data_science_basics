# Merge Data

```python
dm1 = DataFrame({'key':['x-man','zebra','blue','yellow','pink'],'col_1': np.arange(5),'key2':['a','b','c','d','e']})
dm1

```

```python
dm2 = DataFrame({'key':['Queen','zebra','blue','yellow'],'col_2': [1,2,3,4],'col_3':['a','c','c','c']})
dm2
```

```python
pd.merge(dm1,dm2, on='key')
```

```python
pd.merge(dm1,dm2, on='key', how='left')
```

```python
pd.merge(dm2,dm1, on='key', how='left')
```

```python
pd.merge(dm2,dm1, on='key', how='right')
```

```python
pd.merge(dm1,dm2, on='key', how='outer', sort='True')
```
