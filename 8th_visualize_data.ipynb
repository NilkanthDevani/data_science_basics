```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Barplot

```python
df = pd.read_csv('D:/Jupyter/classroom/bar_chart_data.csv')
df
```

```python
#size
plt.figure(figsize=(8,6))
plt.bar(x = df['Brand'], height = df['Cars Listings'], color='midnightblue')
plt.xticks(rotation = 70)
plt.xlabel('number of listing')
plt.title('car',fontsize=25)
```

```python
#size
plt.figure(figsize=(8,6))
s = df.sort_values(by='Cars Listings')
print(s)
plt.barh(y = s['Brand'],
         width = s['Cars Listings']
         
         , color='midnightblue')
# plt.xticks(rotation = 70)
plt.xlabel('number of listing')
plt.title('car',fontsize=25)
```
